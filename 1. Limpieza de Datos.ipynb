## MERGE

```python
import pandas as pd

# Supongamos que tienes tres DataFrames llamados df1, df2 y df3

# Concatenar los DataFrames en uno solo
dataframes = [desempleo1, desempleo2, desempleo3]
df_merged = pd.concat(dataframes)
```

```python
df_merged = df_merged.drop(columns=['Empleo'])
```

```python
nombres_columnas = {
    'Azuay':'AZUAY',
    'Bolívar':'BOLIVAR',
    'Cañar':'CAÑAR',
    'Carchi':'CARCHI',
    'Cotopaxi':'COTOPAXI',
    'Chimborazo':'CHIMBORAZO',
    'El Oro':'EL ORO',
    'Esmeraldas':'ESMERALDAS',
    'Guayas':'GUAYAS',
    'Imbabura':'IMBABURA',
    'Loja':'LOJA',
    'Los Ríos':'LOS RIOS',
    'Manabí':'MANABI',
    'Morona Santiago':'MORONA SANTIAGO',
    'Napo':'NAPO',
    'Pastaza':"PASTAZA",
    'Pichincha':'PICHINCHA',
    'Tungurahua':'TUNGURAHUA',
    'Zamora Chinchipe':'ZAMORA CHINCHIPE',
    'Galápagos':'GALAPAGOS',
    'Sucumbíos':'SUCUMBIOS',
    'Orellana':'ORELLANA',
    'Santo Domingo':'SANTO DOMINGO DE LOS TSACHILAS',
    'Santa Elena':'SANTA ELENA',
    'Quito':'QUITO',
    'Guayaquil':'GUAYAQUIL',
    'Cuenca':'CUENCA',
    'Machala':'MACHALA',
    'Ambato':'AMBATO'
    }
df_merged = df_merged.drop(columns=['Indicador'])
# Renombrar las columnas utilizando el diccionario
df_merged.rename(columns=nombres_columnas, inplace=True)

df_merged
```

```python
columnas_numericas = ['QUITO', 'GUAYAQUIL', 'CUENCA', 'MACHALA', 'AMBATO']

df_merged[columnas_numericas] = df_merged[columnas_numericas].apply(lambda x: x.str.replace('-', 'NaN'))
df_merged[columnas_numericas] = df_merged[columnas_numericas].apply(lambda x: x.str.replace(',', '.')).astype(float)
```

```python
df_merged.dtypes
```

```python
columnas = ['AZUAY', 'BOLIVAR', 'CAÑAR', 'CARCHI', 'COTOPAXI', 'CHIMBORAZO', 'EL ORO', 'ESMERALDAS', 'GUAYAS', 'IMBABURA', 'LOJA', 'LOS RIOS', 'MANABI', 'MORONA SANTIAGO', 'NAPO', 'PASTAZA', 'PICHINCHA', 'TUNGURAHUA', 'ZAMORA CHINCHIPE', 'GALAPAGOS', 'SUCUMBIOS', 'ORELLANA', 'SANTO DOMINGO DE LOS TSACHILAS', 'SANTA ELENA', 'QUITO', 'GUAYAQUIL', 'CUENCA', 'MACHALA', 'AMBATO']

# Imputación por la mediana
df_merged[columnas] = df_merged[columnas].fillna(df_merged[columnas].median())


df_merged
```

```python
df_merged.isnull().sum()
```

```python

# Supongamos que tienes un DataFrame llamado df_merged
# Creamos un DataFrame para las provincias
df_provincias = df_merged[['Año', 'AZUAY', 'BOLIVAR', 'CAÑAR', 'CARCHI', 'COTOPAXI', 'CHIMBORAZO', 'EL ORO', 'ESMERALDAS', 'GUAYAS', 'IMBABURA', 'LOJA', 'LOS RIOS', 'MANABI', 'MORONA SANTIAGO', 'NAPO', 'PASTAZA', 'PICHINCHA', 'TUNGURAHUA', 'ZAMORA CHINCHIPE', 'GALAPAGOS', 'SUCUMBIOS', 'ORELLANA', 'SANTO DOMINGO DE LOS TSACHILAS', 'SANTA ELENA']]
# Derretimos el DataFrame para transformar las columnas de provincias en filas
df_provincias_melted = pd.melt(df_provincias, id_vars=['Año'], var_name='Provincia', value_name='Tasa de Desempleo Provincia')

# Creamos un DataFrame para los cantones
df_cantones = df_merged[['Año', 'QUITO', 'GUAYAQUIL', 'CUENCA', 'MACHALA', 'AMBATO']]
# Derretimos el DataFrame para transformar las columnas de cantones en filas
df_cantones_melted = pd.melt(df_cantones, id_vars=['Año'], var_name='Canton', value_name='Tasa de Desempleo Canton')

# Fusionamos los DataFrames de provincias y cantones en función del año
df_final = pd.merge(df_provincias_melted, df_cantones_melted, on='Año', how='outer')

# Ahora, df_final contiene las columnas de año, desempleo de provincia, desempleo de cantón, provincia y cantón
df_final
```
