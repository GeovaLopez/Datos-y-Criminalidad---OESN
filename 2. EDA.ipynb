```python
#un nuevo dataset df1, limpio y filtrado y concatenado:
### CREANDO UN NUEVO DATASET DF1
Delito = [
"ART. 084 DESAPARICION FORZADA",
"ART. 161 SECUESTRO",
"ART. 162 SECUESTRO EXTORSIVO",
"ART. 185 EXTORSION",
"ART. 151 TORTURA",
"ART. 152 LESIONES",
"ART. 366 TERRORISMO",
"ART. 369 DELINCUENCIA ORGANIZADA",
"ART. 370 ASOCIACION ILICITA",
"ART. 349 GRUPOS SUBVERSIVOS",
"ART. 367 FINANCIACION DEL TERRORISMO",
"ART. 378 MUERTE PROVOCADA POR NEGLIGENCIA DE CONTRATISTA O EJECUTOR DE OBRA",
"ART. 115 HOMICIDIO DE PERSONA PROTEGIDA",
"ART. 144 HOMICIDIO",
"ART. 145 HOMICIDIO CULPOSO",
"ART. 140 ASESINATO",
"ART. 143 SICARIATO",
"ART. 341 TENTATIVA DE ASESINATO CONTRA LA O EL PRESIDENTE DE LA REPUBLICA",
"ART. 085 EJECUCION EXTRAJUDICIAL",
"ART. 146 HOMICIDIO CULPOSO POR MALA PRACTICA PROFESIONAL",
"ART. 122 UTILIZACION DE ARMAS PROHIBIDAS",
"ART. 359 ABUSO DE ARMA DE FUEGO",
"ART. 360 TENENCIA Y PORTE DE ARMAS",
"ART. 189 ROBO",
"ART. 283 ATAQUE O RESISTENCIA",
"ART. 361 ARMAS DE FUEGO,  Y EXPLOSIVOS NO AUTORIZADOS",
"ART. 362 TRAFICO ILICITO DE ARMAS DE FUEGO, ARMAS QUIMICAS, NUCLEARES O BIOLOGICAS"
]

# Filtrar los datos por valores específicos en la columna "Delito"
df1 = df[df["Delito"].isin(Delito)]

# Eliminar las columnas Judicatura, Instancia y Corte
columns_to_drop = ["Judicatura", "Instancia", "Corte", 'Ingreso', 'Resuelta' ,'Tramite', 'Materia', 'TipoAccion'] #no estan en el estudio
df1.drop(columns=columns_to_drop, inplace=True)

# Renombre las columnas Periodo por Año
column_rename = {"Periodo": "Año"}

df1.rename(columns=column_rename, inplace=True)
```

```python
# Convertir la columna "Año" al tipo de dato datetime
df1["Año"] = pd.to_datetime(df1["Año"], format="%d/%m/%Y")

# Extraer el año y asignarlo a una nueva columna
df1["Año"] = df1["Año"].dt.year
```

```python
# DataFrame filtrado
df1.shape
df1
```

```python
soc = pd.read_csv(r"C:\Users\geova\OneDrive\Escritorio\socioeconomico.csv")
soc
```

```python
#DaTa de datos socioeconomicos
data = pd.merge(df1,soc, on=['Año', 'Provincia',"Canton"], how='left')
#data.fillna(0)
data
```

```python
data.isnull().sum()
```

```python
#'Pobreza Provincia' y 'Pobreza Canton' a float64
data['Pobreza Provincia'] = data['Pobreza Provincia'].str.replace('%', '').str.replace(',', '.')
data['Pobreza Provincia'] = data['Pobreza Provincia'].astype(float)
data['Pobreza Canton'] = data['Pobreza Canton'].str.replace('%', '').str.replace(',', '.')
data['Pobreza Canton'] = data['Pobreza Canton'].astype(float)
```

```python
columnas = ['Pobreza Provincia', 'Pobreza Canton']
data[columnas] = data[columnas].fillna(0)
```

```python
columnas = ['Tasa de Desempleo Provincia', 'Tasa de Desempleo Canton', 'Poblacion Provincia', 'Población Canton']
data[columnas] = data[columnas].fillna(data[columnas].mean())
```

### EDA, - correlaciones -

```python
#Estadisticas Descriptivas 
data.describe()  # solo 3 columnas son valores numericos tipo  , voy a eliminar estas columnas...
```

```python
# columnas categóricas en variables dummy (freceuncia encoding)
provincia_frecuencias = data['Provincia'].value_counts().to_dict()
canton_frecuencias = data['Canton'].value_counts().to_dict()
delito_frecuencias = data['Delito'].value_counts().to_dict()

data.hist(figsize= (15,13))
```

```python
# Codificación de frecuencias para variables categóricas
provincia_frecuencias = data['Provincia'].value_counts().to_dict()
canton_frecuencias = data['Canton'].value_counts().to_dict()
delito_frecuencias = data['Delito'].value_counts().to_dict()

data['Provincia'] = data['Provincia'].map(provincia_frecuencias)
data['Canton'] = data['Canton'].map(canton_frecuencias)
data['Delito'] = data['Delito'].map(delito_frecuencias)

# Matriz de correlación
correlation_matrix = data.corr()

# Crear un mapa de calor con seaborn
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
correlation_matrix 
```

```python
from scipy import stats


#chi2 xq son variables nominales y ordinal
# creo una tabla de contingencia
contingency_table = pd.crosstab(df1['Delito'], df1['Año'])
# calculo chi2
chi2, p_value, dof, expected_values = stats.chi2_contingency(contingency_table)

print("Chi2:", chi2)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
#print("Expected values:")
#print(expected_values) ## no hay evidencia para afirmar que existe relación entre Delito y Año, osea los delitos no tienen nada que ver con el año
```

```python
# existe correlación entre el delito y la provincia?

#  tabla de contingencia
contingency_table = pd.crosstab(df1['Delito'], df1['Provincia'])
hi2, p_value, dof, expected_values = stats.chi2_contingency(contingency_table)

print("Chi2:", chi2)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
# no hay relación osea que el delito no tiene nada que ver con la provincia 
```
